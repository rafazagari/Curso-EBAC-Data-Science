# Cálculo 

1- Calcule de forma numérica a integral da função $x^3$ avaliada entre os pontos $1$ e $2$. Identifique o número de retângulos que você usou.  
2- Calcule essa área de forma analítica.

```python
# Imports
import numpy as np
import matplotlib.pyplot as plt

# Calculo numérico

# Função para ser integrada:
def f(x):
    return x**3

# Limites da integrada:
l1 = 1
l2 = 2

# Quantidade retangulos da integral:
n_retangulos = 1000

# Pontos de x e y:
x_valores = np.linspace(l1, l2, n_retangulos+1)
y_valores = f(x_valores)

# Intervalos dos pontos
x_intervalos = (l2 - l1)/n_retangulos

# Área de cada retangulo:
area_retangulos = x_intervalos * y_valores

# Soma das areas:
integral = np.sum(area_retangulos)

# Plotagem 
plt.plot(x_valores, y_valores, label='f(x) = x^3', color='red')
plt.bar(x_valores, y_valores, width=x_intervalos, alpha=0.3, align='edge', label='Retângulos')
plt.title('Aproximação Numérica da Integral')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()

# Resultados
print(f"Aproximação Numérica da Integral: {integral}")
print(f"Número de Retângulos Usados: {n_retangulos}")
```

# Calculo de forma analitica

\begin{equation}
\int_{1}^{2} x^3 \,dx = \left[\frac{1}{4}x^4\right]_{1}^{2} 
= \frac{1}{4}(2)^4 - \frac{1}{4}(1)^4 
= \frac{1}{4}(16) - \frac{1}{4}(1)
= 4 - \frac{1}{4}
= \frac{15}{4}
\end{equation}

\begin{equation}

\end{equation}
