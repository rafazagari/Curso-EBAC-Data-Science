# Probabilidade 

```python
import numpy as np
import matplotlib.pyplot as plt

from scipy import stats
```

1. Uma máquina embala ração para cachorro em pacotes de 15 kg, mas cada pacote produzido acaba tendo um peso aleatório, com distribuição Normal, com média 15kg, mas com desvio padrão 0,1 kg. Obtenha a probabilidade de que um pacote selecionado aleatoriamente dessa produção tenha mais que 15,2 kg.

```python
# Parametros para calculo
media = 15
desv_pad = 0.1

# Calculo da probabilidade de um pacote pesar mais que 15.2 kg
prob = 1 - stats.norm.cdf(15.2, media, desv_pad)
print("A probabilidade de que um pacote selecionado aleatoriamente tenha mais que 15,2 kg é de: {:.2%}".format(prob))
```

2. Da máquina descrita acima, obtenha o percentil de ordem 95% para o peso dos pacotes. Ou seja, o valor tal que a probabilidade de um pacote ter peso menor que este seja de 95%.

```python
# Parametros para calculo

percentil_95 = 0.95

# Calculo do percentil 95%
peso_percentil_95 = stats.norm.ppf(percentil_95, loc=media, scale=desv_pad)
print("O percentil de ordem 95% para o peso dos pacotes é de: {:.2f} kg".format(peso_percentil_95))
```

3. Suponha duas variáveis aleatórias $Y_1$ e $Y_2$, a primeira com distribuição Normal Padrão (média=0, variância=1), a segunda com distribuição *t-studdent* com 5 graus de liberdade (também padrão). Calcule, para cada uma, a probabilidade de ser menor que cada um dos valores contidos no objeto $Y$ na célula abaixo:

```python
Y=np.linspace(-6, 6, 200)
# Seu código começa aqui

# Distribuição Normal Padrão (Y1)
prob_Y1 = stats.norm.cdf(Y, loc=0, scale=1)

# Distribuição t-Student com 5 graus de liberdade (Y2)
prob_Y2 = stats.t.cdf(Y, df=5)

# Exibir as probabilidades para cada valor em Y
for i in range(len(Y)):
    print("Para Y =", Y[i])
    print("Probabilidade para Y1 (Normal Padrão):", prob_Y1[i])
    print("Probabilidade para Y2 (t-Student):", prob_Y2[i])
    print()
```

4. Faça uma visualização gráfica dos dados do item anterior.

```python
# Para realizar o plot dos gráficos no matplotlib, primeiro encontro a densidade de probabilidade (PDF) das variáveis.
# Distribuição Normal Padrão (Y1)
pdf_Y1 = stats.norm.pdf(Y, loc=0, scale=1)

# Distribuição t-Student com 5 graus de liberdade (Y2)
pdf_Y2 = stats.t.pdf(Y, df=5)

# Plotar as distribuições
plt.figure(figsize=(10, 6))
plt.plot(Y, pdf_Y1, label='Y1 (Normal Padrão)')
plt.plot(Y, pdf_Y2, label='Y2 (t-Student)')
plt.title('Distribuições de Probabilidade de Y1 e Y2')
plt.xlabel('Valores de Y')
plt.ylabel('Densidade de Probabilidade')
plt.legend()
plt.grid(True)
plt.show()
```

5. 3. Suponha duas variáveis aleatórias $Y_1$ e $Y_2$, a primeira com distribuição Normal Padrão (média=0, variância=1), a segunda com distribuição *t-studdent* com 5 graus de liberdade (também padrão). Calcule, para cada uma, o percentil (a inversa da *CDF*) para cada um dos valores do objeto *P* abaixo:

```python
P=np.linspace(0, 1, 20)
# Seu código começa aqui
# Distribuição Normal Padrão (Y1)
percentis_Y1 = stats.norm.ppf(P, loc=0, scale=1)

# Distribuição t-Student com 5 graus de liberdade (Y2)
percentis_Y2 = stats.t.ppf(P, df=5)

# Exibir os percentis para cada valor em P
for i in range(len(P)):
    print("Para P =", P[i])
    print("Percentil para Y1 (Normal Padrão):", percentis_Y1[i])
    print("Percentil para Y2 (t-Student):", percentis_Y2[i])
    print()
```

6. Faça um gráfico com os dados do item anterior.

```python
# Plotar os percentis
plt.figure(figsize=(10, 6))
plt.plot(P, percentis_Y1, label='Y1 (Normal Padrão)')
plt.plot(P, percentis_Y2, label='Y2 (t-Student)')
plt.title('Percentis de Y1 e Y2')
plt.xlabel('Percentil')
plt.ylabel('Valor')
plt.legend()
plt.grid(True)
plt.show()
```
