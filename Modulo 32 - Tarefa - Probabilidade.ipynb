# Probabilidade 

```python
import numpy as np
import matplotlib.pyplot as plt

from scipy import stats
```

1. Uma mquina embala rao para cachorro em pacotes de 15 kg, mas cada pacote produzido acaba tendo um peso aleatrio, com distribuio Normal, com mdia 15kg, mas com desvio padro 0,1 kg. Obtenha a probabilidade de que um pacote selecionado aleatoriamente dessa produo tenha mais que 15,2 kg.

```python
# Parametros para calculo
media = 15
desv_pad = 0.1

# Calculo da probabilidade de um pacote pesar mais que 15.2 kg
prob = 1 - stats.norm.cdf(15.2, media, desv_pad)
print("A probabilidade de que um pacote selecionado aleatoriamente tenha mais que 15,2 kg  de: {:.2%}".format(prob))
```

2. Da mquina descrita acima, obtenha o percentil de ordem 95% para o peso dos pacotes. Ou seja, o valor tal que a probabilidade de um pacote ter peso menor que este seja de 95%.

```python
# Parametros para calculo

percentil_95 = 0.95

# Calculo do percentil 95%
peso_percentil_95 = stats.norm.ppf(percentil_95, loc=media, scale=desv_pad)
print("O percentil de ordem 95% para o peso dos pacotes  de: {:.2f} kg".format(peso_percentil_95))
```

3. Suponha duas variveis aleatrias $Y_1$ e $Y_2$, a primeira com distribuio Normal Padro (mdia=0, varincia=1), a segunda com distribuio *t-studdent* com 5 graus de liberdade (tambm padro). Calcule, para cada uma, a probabilidade de ser menor que cada um dos valores contidos no objeto $Y$ na clula abaixo:

```python
Y=np.linspace(-6, 6, 200)
# Seu cdigo comea aqui

# Distribuio Normal Padro (Y1)
prob_Y1 = stats.norm.cdf(Y, loc=0, scale=1)

# Distribuio t-Student com 5 graus de liberdade (Y2)
prob_Y2 = stats.t.cdf(Y, df=5)

# Exibir as probabilidades para cada valor em Y
for i in range(len(Y)):
    print("Para Y =", Y[i])
    print("Probabilidade para Y1 (Normal Padro):", prob_Y1[i])
    print("Probabilidade para Y2 (t-Student):", prob_Y2[i])
    print()
```

4. Faa uma visualizao grfica dos dados do item anterior.

```python
# Para realizar o plot dos grficos no matplotlib, primeiro encontro a densidade de probabilidade (PDF) das variveis.
# Distribuio Normal Padro (Y1)
pdf_Y1 = stats.norm.pdf(Y, loc=0, scale=1)

# Distribuio t-Student com 5 graus de liberdade (Y2)
pdf_Y2 = stats.t.pdf(Y, df=5)

# Plotar as distribuies
plt.figure(figsize=(10, 6))
plt.plot(Y, pdf_Y1, label='Y1 (Normal Padro)')
plt.plot(Y, pdf_Y2, label='Y2 (t-Student)')
plt.title('Distribuies de Probabilidade de Y1 e Y2')
plt.xlabel('Valores de Y')
plt.ylabel('Densidade de Probabilidade')
plt.legend()
plt.grid(True)
plt.show()
```

5. 3. Suponha duas variveis aleatrias $Y_1$ e $Y_2$, a primeira com distribuio Normal Padro (mdia=0, varincia=1), a segunda com distribuio *t-studdent* com 5 graus de liberdade (tambm padro). Calcule, para cada uma, o percentil (a inversa da *CDF*) para cada um dos valores do objeto *P* abaixo:

```python
P=np.linspace(0, 1, 20)
# Seu cdigo comea aqui
# Distribuio Normal Padro (Y1)
percentis_Y1 = stats.norm.ppf(P, loc=0, scale=1)

# Distribuio t-Student com 5 graus de liberdade (Y2)
percentis_Y2 = stats.t.ppf(P, df=5)

# Exibir os percentis para cada valor em P
for i in range(len(P)):
    print("Para P =", P[i])
    print("Percentil para Y1 (Normal Padro):", percentis_Y1[i])
    print("Percentil para Y2 (t-Student):", percentis_Y2[i])
    print()
```

6. Faa um grfico com os dados do item anterior.

```python
# Plotar os percentis
plt.figure(figsize=(10, 6))
plt.plot(P, percentis_Y1, label='Y1 (Normal Padro)')
plt.plot(P, percentis_Y2, label='Y2 (t-Student)')
plt.title('Percentis de Y1 e Y2')
plt.xlabel('Percentil')
plt.ylabel('Valor')
plt.legend()
plt.grid(True)
plt.show()
```
