#### 2. Média móvel

1. Faça uma média móvel  em 90 dias dos valores de fechamento ajustados dos ativos que você escolheu
2. Utilizando a função ```rolling()```, calcule o desvio padrão desses ativos ao longo do tempo, utilizando a mesma janela móvel do item acima
2. Monte um gráfico de linhas para as médias móveis dos ativos da sua base
4. Monte um gráfico de linhas para os desvios padrão em janela móvel
5. Compare os dois gráficos que você fez com o da tarefa anterior. Veja se eles apontam conclusões semelhantes quanto aos períodos de maior volatilidade (variação) dos ativos.

````python
# Para facilitar os gráficos e calculos posteriores, irei realizar algumas alterações no DF
data_filt_fechamento = (
    data_filt[['Adj Close, AMD', 'Adj Close, MGLU3.SA', 'Adj Close, WEGE3.SA']] #Limpeza do dataset e mantendo somente as colunas "Adj Close"
    # .set_index('Data') #Definindo a coluna 'Date' como index
     )

data_filt_fechamento.head(5)
````

````python
#Lista com os ativos
ativos = ['AMD', 'MGLU3.SA', 'WEGE3.SA']

#Calculo das médias móveis e preparação do DF para gráfico
data_med = data_filt_fechamento.rolling(90).mean()
data_med_graf = data_med.stack().reset_index()
data_med_graf.columns = ['data', 'ativos', 'media movel']

data_med_graf.head(5)
````

````python
# Calculo do desvio padrão e preparação do DF para gráfico
data_desv = data_filt_fechamento.rolling(10).std()
data_desv_graf = data_desv.stack().reset_index()
data_desv_graf.columns = ['data', 'ativos', 'desvio padrao']

data_desv_graf.head(5)
````

````python
#Gráfico dos valores médios de ativos
fig_1 = px.line(data_med_graf, x= "data", y="media movel", color='ativos')
fig_1.update_layout(title='Valor Medio de Ativos')
fig_1.update_xaxes(title_text="Data")
fig_1.update_yaxes(title_text="Média móvel")

#Gráfico do desvio padrão de ativos
fig_2 = px.line(data_desv_graf, x= "data", y="desvio padrao", color='ativos')
fig_2.update_layout(title='Desvio Padrão de Ativos')
fig_2.update_xaxes(title_text="Data")
fig_2.update_yaxes(title_text="Desvio padrão")

fig_1.show()
fig_2.show()
````

Em comparação com o da atividade anterior é possível observar que o ativo com maior volatilidade, o ativo AMD, também apresentou grandes variações no desvio padrão e média móvel oscilante.

#### 3. Gráfico de *candlestick*
1. Selecione o período dos últimos 60 dias (corridos) para um dos ativos da sua base
2. Monte um gráfico de *candlestick* para esses dados.

````python
#Preparando o DF para as operações e modificações
data_filt.index = pd.to_datetime(data_filt.index)

#Selecionando as colunas desejadas par o gráfico
colunas_desejadas = ['Close, AMD', 'High, AMD', 'Low, AMD', 'Open, AMD']
amd_filt = data_filt[colunas_desejadas]

#Separando as datas dos últimos 60 dias
amd = amd_filt.tail(60)
amd.head(3)
````

````python
# Crie um gráfico de candlestick
candlestick = go.Candlestick(x=amd.index,
                             open=amd['Open, AMD'],
                             close=amd['Close, AMD'],
                             high=amd['High, AMD'],
                             low=amd['Low, AMD'],
                             name='AMD',
                             showlegend=True)

# Crie um layout para o gráfico
layout = go.Layout(title='Gráfico de Candlestick - AMD',
                   xaxis=dict(title='Data'),
                   yaxis=dict(title='Preço'),
                   showlegend=True)

# Crie uma figura que contém o gráfico de candlestick e o layout
fig = go.Figure(data=[candlestick], layout=layout)

# Exiba o gráfico
fig.show()
````
