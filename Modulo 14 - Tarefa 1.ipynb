```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set()

# Definindo a função de platar os gráficos
def plota_pivot_table(df, value, index, func, ylabel, xlabel, opcao='unstack', meses=None):
    if opcao == 'nada':
      # Definindo o filtro para selecionar os meses de março, abril, maio, junho, dezembro
        if meses:
            filtered_df = df[df['DTNASC'].str[:7].isin(meses)]
            pd.pivot_table(filtered_df, values=value, index=index, aggfunc=func).plot(figsize=[15, 5])
    elif opcao == 'unstack':
        pd.pivot_table(df, values=value, index=index, aggfunc=func).unstack().plot(figsize=[15, 5])
    elif opcao == 'sort':
        pd.pivot_table(df, values=value, index=index, aggfunc=func).sort_values(value).plot(figsize=[15, 5])
    plt.ylabel(ylabel)
    plt.xlabel(xlabel)
    return None

# Lendo os dados da database
sinasc = pd.read_csv('/content/SINASC_RO_2019.csv')

# Definindo os meses desejados
meses_desejados = ['2019-03', '2019-04', '2019-05', '2019-06', '2019-12']

# Chamando a função testar o plot dos gráficos
plota_pivot_table(sinasc, value='PESO', index=['ESCMAE'], func='mean', ylabel='Média Alfabetização da Mãe', xlabel='Idade da Mãe', opcao='unstack')

plt.show()
```
