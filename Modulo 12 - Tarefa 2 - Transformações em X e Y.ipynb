# Regressão 01 - transformações em X e Y

### 1. Melhorando o ajuste no percentual de gorjetas

a. Vamos partir do modelo feito na última tarefa, o que relaciona ```tip_pct``` e ```net_bill```. Carregue a base, os pacotes e reproduza este modelo aqui.

````python
import pandas as pd
import seaborn as sns
from seaborn import load_dataset

import matplotlib.pyplot as plt

import numpy as np

import statsmodels.api as sm
import statsmodels.formula.api as smf

tips = sns.load_dataset("tips")
tips['tip_pct'] = tips['tip'] / (tips['total_bill'] - tips['tip'])
tips['net_bill'] = tips['total_bill'] - tips['tip']
tips.head()

modelo_tip_pct = smf.ols('tip_pct ~ net_bill', data = tips).fit()
modelo_tip_pct.summary()
````

### 2. ajuste o modelo de ```tip_pct``` em função do logaritmo de ```net_bill```.

````python
modelo_log_y = smf.ols('tip_pct ~ np.log(net_bill)', data = tips).fit()
print(modelo_log_y.summary())
````

###  3. ajuste o modelo de ```tip_pct``` em função de um polinômio de segundo grau em ```net_bill```.

````python
modelo_power_y = smf.ols('tip_pct ~ np.power(net_bill,2)', data = tips).fit()
print(modelo_power_y.summary())
````

###  4. ajuste um modelo no log do ```tip_pct``` em função de ```net_bill```. Calcule o $R^2$ também em ```tip_pct```

````python
modelo_log_x = smf.ols('np.log(tip_pct) ~ net_bill', data = tips).fit()
print(modelo_log_x.summary())
````

###  5. ajuste um modelo no log do ```tip_pct``` em função do log de ```net_bill```. Calcule o $R^2$ também em ```tip_pct```.

````python
modelo_log_xy = smf.ols('np.log(tip_pct) ~ np.log(net_bill)', data = tips).fit()
print(modelo_log_xy.summary())
````

###  6. qual modelo resultou no melhor ajuste?

O melhor modelo, que apresentou um R² de 0,293, foi o que ajustei no log do tip_pct e log de net_bill.
