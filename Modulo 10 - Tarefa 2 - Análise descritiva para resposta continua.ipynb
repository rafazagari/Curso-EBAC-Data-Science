````python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df_renda = pd.read_csv('previsao_de_renda.csv')
df_renda.head()
````

````python
sns.pointplot(x = "posse_de_imovel", y = "renda", data = df_renda, dodge = True)
````
